```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.sans-serif'] = 'SimHei' ## 设置中文显示
plt.rcParams['axes.unicode_minus'] = False ## 正常显示负号
# 设置打印内容全部输出
np.set_printoptions(threshold=np.inf)
data = np.load('./国民经济核算季度数据.npz', allow_pickle=True)
# data.files
values = data['values']
columns = data['columns']
label1=['第一产业','第二产业','第三产业']
label2=['农业','工业','建筑','批发','交通','餐饮','金融','房地产','其他']
color_list = [
    (0.9, 0.1, 0.1), (0.1, 0.9, 0.1), (0.1, 0.1, 0.9), (0.9, 0.9, 0.1), (0.9, 0.1, 0.9),
    (0.1, 0.9, 0.9), (0.3, 0.6, 0.9), (0.6, 0.3, 0.9), (0.9, 0.3, 0.6)]
```

```python
columns
```

```python
fig = plt.figure(figsize=(12, 12), dpi=100)
# 子图1
fig.add_subplot(2, 2, 1)
plt.bar(range(3), values[0, 3:6], width=0.6, color=['r', 'g', 'b'])
plt.xlabel('产业')
plt.ylabel('生产总值(亿元)')
plt.title('2000年第一季度国民生产总值产业构成柱状图')
plt.xticks(range(3), label1)
for a, b in zip(range(3), values[0, 3:6]):
    plt.text(a, b, '%.2f'%b, ha='center', va='bottom', fontsize=12)
# 子图2
fig.add_subplot(2, 2, 2)
plt.bar(range(3), values[-1, 3:6], width=0.6, color=['r', 'g', 'b'])
plt.xlabel('产业')
plt.ylabel('生产总值(亿元)')
plt.title('2017年第一季度国民生产总值产业构成柱状图')
plt.xticks(range(3), label1)
for a, b in zip(range(3), values[-1, 3:6]):
    plt.text(a, b, '%.2f'%b, ha='center', va='bottom', fontsize=12)
# 子图3
fig.add_subplot(2, 2, 3)
plt.bar(range(9), values[0, 6:], width=0.6, color=color_list)
plt.xlabel('行业')
plt.ylabel('生产总值(亿元)')
plt.title('2000年第一季度国民生产总值行业构成柱状图')
plt.xticks(range(9), label2)
for a, b in zip(range(9), values[0, 6:]):
    plt.text(a, b, '%.2f'%b, ha='center', va='bottom', fontsize=12)
# 子图4
fig.add_subplot(2, 2, 4)
plt.bar(range(9), values[-1, 6:], width=0.6, color=color_list)
plt.xlabel('行业')
plt.ylabel('生产总值(亿元)')
plt.title('2017年第一季度国民生产总值行业构成柱状图')
plt.xticks(range(9), label2)
for a, b in zip(range(9), values[-1, 6:]):
    plt.text(a, b, '%.2f'%b, ha='center', va='bottom', fontsize=12)
plt.show()
```

```python
explode1=[0.01,0.02,0.01]
explode2=[0.01,0.01,0.01,0.01,0.01,0.01,0.01,0.01,0.01]
fig2 = plt.figure(figsize=(12, 12), dpi=100)
# 子图1
fig2.add_subplot(2, 2, 1)
plt.pie(x=values[0, 3:6], labels=label1, explode=explode1, autopct='%4.1f%%')
# 子图2
fig2.add_subplot(2, 2, 2)
plt.pie(x=values[-1, 3:6], labels=label1, explode=explode1, autopct='%4.1f%%')
# 子图3
fig2.add_subplot(2, 2, 3)
plt.pie(x=values[0, 6:], labels=label2, explode=explode2, autopct='%4.1f%%')
# 子图4
fig2.add_subplot(2, 2, 4)
plt.pie(x=values[-1, 6:], labels=label2, explode=explode2, autopct='%4.1f%%')
plt.show()
```
