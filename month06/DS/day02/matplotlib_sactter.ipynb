```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.sans-serif'] = 'SimHei' ## 设置中文显示
plt.rcParams['axes.unicode_minus'] = False ## 正常显示负号
```

```python
# 设置打印内容全部输出
np.set_printoptions(threshold=np.inf)
data = np.load('./国民经济核算季度数据.npz', allow_pickle=True)
# data.files
values = data['values']
columns = data['columns']
```

```python
# 绘制2000年-2007年各产业国民经济生产总值散点图
plt.figure(figsize=(10,8))
# 第一产业
plt.scatter(x=values[:, 0], y=values[:, 3], marker='o', c=values[:, 3], cmap=plt.cm.viridis)
# 第二产业
plt.scatter(x=values[:, 0], y=values[:, 4], marker='D', c=values[:, 4], cmap=plt.cm.Reds)
# 第三产业
plt.scatter(x=values[:, 0], y=values[:, 5], marker='s', c=values[:, 5], cmap=plt.cm.Greens)
# 添加标签
plt.xlabel('序号')
plt.ylabel('生产总值')
# 标题
plt.title('2000-2007年各产业国民经济生产总值散点图')
# 添加图例
plt.legend(['第一产业', '第二产业', '第三产业'])
plt.savefig('scatter01.png')
plt.show()
```

```python
# 绘制2000年-2017年各产业及各行业国民生产总值折线图
label1=['第一产业','第二产业','第三产业']
label2=['农业','工业','建筑','批发','交通','餐饮','金融','房地产','其他']
# 创建画布
fig2 = plt.figure(figsize=(15, 10), dpi=100)
# 创建子图1
fig2.add_subplot(2, 1, 1)
# 绘制三大产业折线图
plt.plot(
        values[:, 0], values[:, 3], 'b-',
        values[:, 0], values[:, 4], 'r-.',
        values[:, 0], values[:, 5], 'g--')
plt.ylabel('生产总值(亿元)')
plt.title('2000年-2017年各产业国民生产总值折线图')
# 创建子图2
fig2.add_subplot(2, 1, 2)
# 绘制九大行业折线图
plt.plot(
        values[:, 0], values[:, 6], 'b-',
        values[:, 0], values[:, 7], 'r-.',
        values[:, 0], values[:, 8], 'g--',
        values[:, 0], values[:, 9], 'y-',
        values[:, 0], values[:, 10], 'c-.',
        values[:, 0], values[:, 11], 'k--',
        values[:, 0], values[:, 12], 'm-',
        values[:, 0], values[:, 13], 'g-.',
        values[:, 0], values[:, 14], 'b--',)
plt.xlabel('年份')
plt.ylabel('生产总值')
plt.legend(label2)
plt.title('2000年-2017年各行业国民生产总值折线图')
# plt.xticks(range(0, 70, 4), values[:, 1][::4], rotation=45)
plt.xticks(range(0, 70, 4), values[range(0, 70, 4), 1], rotation=45)
plt.savefig('plot02.png')
# 显示图形
plt.show()
```
