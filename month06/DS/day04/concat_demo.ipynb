```python
import numpy as np
import pandas as pd
```

```python
# 定义数据
df1 = pd.DataFrame(np.zeros((3, 4)), columns=list('abcd'))
df2 = pd.DataFrame(np.ones((3, 4)), columns=list('abcd'))
df3 = pd.DataFrame(np.ones((3, 4)) * 2, columns=list('abcd'))
```

```python
df1
```

```python
res = pd.concat(objs=[df1, df2, df3], axis=0, ignore_index=True)
res
```

```python
df4 = pd.DataFrame(np.ones((3, 4)) * 0, columns=['a', 'b', 'c', 'd'], index=[1, 2, 3])
df5 = pd.DataFrame(np.ones((3, 4)) * 1, columns=['b', 'c', 'd', 'e'], index=[2, 3, 4])
```

```python
df4
```

```python
df5
```

```python
# 纵向合并, 求并集
res2 = pd.concat(objs=[df4, df5], axis=0, join='outer', ignore_index=False, keys=['df4', 'df5'],
                names=['df_name', 'row_id'])
res2
```

```python
# 纵向合并, 求交集
res3 = pd.concat(objs=[df4, df5], axis=0, join='inner', ignore_index=False, keys=['df4', 'df5'],
                names=['df_name', 'row_id'])
res3
```

```python
# 横向合并, 求交集
res4 = pd.concat(objs=[df4, df5], axis=1, join='inner', ignore_index=False, keys=['df4', 'df5'],
                names=['df_name', 'row_id'])
res4
```

```python
# 横向合并, 求并集
res4 = pd.concat(objs=[df4, df5], axis=1, join='outer', ignore_index=False, keys=['df4', 'df5'],
                names=['df_name', 'row_id'])
res4
```

```python
# append函数: 纵向堆叠
df6 = pd.DataFrame(np.ones((3, 4)) * 2, columns=list('abcd'))
df7 = pd.DataFrame(np.ones((3, 4)) * 4, columns=list('abcd'))
```
